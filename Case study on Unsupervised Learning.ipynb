# * k means clustering

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init ='k-means++', random_state = 42)
    kmeans.fit(data1)
    wcss.append(kmeans.inertia_)
    print(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=i, init ='k-means++', random_state = 42)
y_kmeans =kmeans.fit_predict(data1)
type(data)
y_kmeans
```

```python
from sklearn.cluster import KMeans
model_kmeans = KMeans(n_clusters =3)
model_kmeans.fit(data1)
labels =model_kmeans.predict(data1)
print(labels)
```

```python
al = data1['Alcohol']
tp = data1['Total_Phenols']
plt.scatter(al,tp,c=labels)
```

# * Hierarchial clustering

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(data1, method ='ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Eucledian distances')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 3, affinity= 'euclidean', linkage='ward')
y_hc = hc.fit_predict(data1)
```

```python
y_hc
```

```python
plt.title("Alcohol and Total_Phenols")
plt.scatter(data["Alcohol"], data["Total_Phenols"], c = y_hc)
plt.xlabel("Alchol")
plt.ylabel("Total_Phenols")
plt.show()
```

# PCA

```python
# scaling the data
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
s_data = scalar.fit_transform(data1)
s_data = pd.DataFrame(s_data, columns = data1.columns)
s_data.describe()                       
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components= 0.95)
pca.fit(s_data)
data_pca = pca.transform(s_data)
print(f'Number of Principal Components to explain 95% variance = {pca.n_components_}')
pca.explained_variance_ratio_
```
